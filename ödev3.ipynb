```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax  = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
a = cv2.imread('Giraffe.jpg')
giraf = cv2.cvtColor(a,cv2.COLOR_BGR2RGB)
plt.imshow(giraf)
```

```python
giraf_black = cv2.cvtColor(a,cv2.COLOR_BGR2GRAY)
ret , th1 = cv2.threshold(giraf_black,127,255,cv2.THRESH_BINARY)
```

```python
display_img(th1,cmap='gray')
```

```python
giraf_hsv = cv2.cvtColor(a,cv2.COLOR_BGR2HSV)
plt.imshow(giraf_hsv)
```

```python
old_array = np.ones([4,4])
array = old_array / 10
array
```

```python
filt = cv2.filter2D(giraf,-1,array)
display_img(giraf)
```

```python

sobely = cv2.Sobel(a,cv2.CV_64F,1,0,ksize = 5)
display_img(sobely,cmap='gray')
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([giraf],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
plt.title('histo')
plt.show()
```
